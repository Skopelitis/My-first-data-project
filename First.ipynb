```python
# Simple Sales Data Analysis
import pandas as pd
import matplotlib.pyplot as plt

# Sample sales data (no file needed!)
data = {
    'Product': ['Laptop', 'Mouse', 'Keyboard', 'Monitor', 'Headphones'],
    'Units_Sold': [45, 120, 78, 32, 95],
    'Price_Per_Unit': [800, 25, 60, 200, 50]
}

df = pd.DataFrame(data)

# Calculate total revenue per product
df['Total_Revenue'] = df['Units_Sold'] * df['Price_Per_Unit']

# Show the data
print("Sales Data Overview:")
print(df)

# Key business metrics
print("\n Key Metrics:")
print(f"Total units sold: {df['Units_Sold'].sum()}")
print(f"Total revenue: ${df['Total_Revenue'].sum():,}")
print(f"Best selling product: {df.loc[df['Units_Sold'].idxmax(), 'Product']} ({df['Units_Sold'].max()} units)")
print(f"Highest revenue product: {df.loc[df['Total_Revenue'].idxmax(), 'Product']} (${df['Total_Revenue'].max():,})")

# Bar chart - Units sold
plt.figure(figsize=(8, 4))
plt.bar(df['Product'], df['Units_Sold'], color='steelblue')
plt.title('Units Sold by Product', fontsize=14)
plt.ylabel('Units Sold')
plt.xlabel('Product')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Bar chart - Revenue
plt.figure(figsize=(8, 4))
plt.bar(df['Product'], df['Total_Revenue'], color='coral')
plt.title('Total Revenue by Product', fontsize=14)
plt.ylabel('Revenue ($)')
plt.xlabel('Product')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

print("\n Found top performer by revenue.")
```
